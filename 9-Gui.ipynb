```python
可变变量通过set()和get()方法设置和获取
由于Tkinter的组织模块，在创建顶层窗口之前，无法创建StringVar
from Tkinter import *
window = Tk()
data = StringVar()
data.set("Data to display")
label = Label(window, textvariable=data)
label.pack()
window.mainloop()

```

```python
Text

多行文本的输入
from Tkinter import *
def cross(text):                                             
    text.insert(INSERT, 'X')
window = Tk()
frame = Frame(window)
frame.pack()
text = Text(frame, height=3, width=10)
text.pack()
button = Button(frame, text="Add", command=lambda: cross(text))
button.pack()
window.mainloop()

```

```python
CheckButton

from Tkinter import *
window = Tk()
frame = Frame(window)
frame.pack()
red = IntVar()                
green = IntVar()
blue = IntVar()

for (name, var) in (('R', red), ('G', green), ('B', blue)):                 #3个
    check = Checkbutton(frame, text=name, variable=var)
    check.pack(side='left')
def recolor(widget, r, g, b):
    color = '#'
    for var in (r, g, b):
        color += 'FF' if var.get() else '00'
    widget.config(bg=color)

label = Label(frame, text='[       ]')
button = Button(frame, text='update',
                command=lambda: recolor(label, red, green, blue))
button.pack(side='left')
label.pack(side='left')
window.mainloop()

```

```python
Menu

from Tkinter import *
import tkFileDialog as dialog
                                                        
def save(root, text):
  data = text.get('0.0', END)
  filename = dialog.asksaveasfilename(
      parent=root,
      filetypes=[('Text', '*.txt')],
      title='Save as...')
  writer = open(filename, 'w')
  writer.write(data)
  writer.close()

def quit(root):
  root.destroy()

window = Tk()
text = Text(window)
text.pack()

menubar = Menu(window)
filemenu = Menu(menubar)
filemenu.add_command(label='Save', command=lambda : save(window, text))
filemenu.add_command(label='Quit', command=lambda : quit(window))

menubar.add_cascade(label = 'File', menu=filemenu)
window.config(menu=menubar)

window.mainloop()

```

```python
MVC

模型model：存储数据
视图view：展示
控制器controller：更新应用程序的模型，触发相应的视图发生变化


MVC
from Tkinter import *
def click():                                                    # The controller.
    counter.set(counter.get() + 1)
if __name__ == '__main__':
    window = Tk()
    counter = IntVar()                                     # The model.
    counter.set(0)
    frame = Frame(window)                             # The views.
    frame.pack()
    button = Button(frame, text="Click", command=click)
    button.pack()
    label = Label(frame, textvariable=counter)
    label.pack()
    window.mainloop()


赋给command的函数不能带有任何参数，只能使用全局变量global variable

```

```python
Lambda


没有名字的单行函数
(lambda x: 2*x)(3)

from Tkinter import *
window = Tk()
counter = IntVar()
counter.set(0)
def click(var, value):
    var.set(var.get() + value)
frame = Frame(window)
frame.pack()
button = Button(frame, text="Up", command=lambda: click(counter, 1))
button.pack()
button = Button(frame, text="Down", command=lambda: click(counter, -1))
button.pack()
label = Label(frame, textvariable=counter)
label.pack()
window.mainloop()


字体
from Tkinter import *
window = Tk()
button = Button(window, text="Hello", font=("Courier", 14, "bold italic"))
button.pack()
window.mainloop()


颜色
from Tkinter import *
window = Tk()
button = Label(window, text="Hello", bg="green", fg="white")
button.pack()
window.mainloop()


布局
from Tkinter import *
window = Tk()
frame = Frame(window)
frame.pack()
label = Label(frame, text="Name")
label.pack(side="left")
entry = Entry(frame)
entry.pack(side="left")
window.mainloop()


布局管理器grid
from Tkinter import *
window = Tk()
frame = Frame(window)
frame.pack()
label = Label(frame, text="Name:")
label.grid(row=0, column=0)
entry = Entry(frame)
entry.grid(row=1, column=1)
window.mainloop()



grid()的参数
row                            从0开始
column                      从0开始
rowspan                    占据的行数，默认1
columnspan               占据的列数，默认1
```

```python
拾色器，根据文本框输入更新一段文本的颜色
from Tkinter import *

def change(widget, colors):
    new_val = '#'
    for name in ('red', 'green', 'blue'):
        new_val += colors[name].get()      #获取用户输入
    widget['bg'] = new_val                       #竟然是字典

window = Tk()
frame = Frame(window)
frame.pack()

colors = {}
for (name, col) in (('red', '#FF0000'),
                    ('green', '#00FF00'),
                    ('blue', '#0000FF')):
    colors[name] = StringVar()
    colors[name].set('00')
    entry = Entry(frame, textvariable=colors[name], bg=col, fg="white")
    entry.pack()

current = Label(frame, text='     ', bg='#FFFFFF')
current.pack()

update = Button(frame, text='Update', command=lambda: change(current, colors))
update.pack()

mainloop()


面向对象GUIpackage
oogui.py
```
